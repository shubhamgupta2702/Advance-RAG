```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
document = [
  "Generative AI systems increasingly rely on Retrieval-Augmented Generation (RAG) to ground outputs in external knowledge sources, improving factual accuracy and contextual relevance.",
  "At the core of RAG pipelines are embeddings, which transform text into dense vector representations that capture semantic meaning across high-dimensional space.",
  "These embeddings enable efficient similarity search over large corpora, allowing models to retrieve the most relevant documents at inference time.",
  "Transformer architectures, with their attention mechanisms, play a central role in both generating embeddings and producing coherent responses.",
  "By integrating retrieval with generation, RAG systems mitigate hallucination issues commonly associated with standalone generative models.",
  "Fine-tuning embedding models and optimizing vector databases are critical for achieving low-latency, high-precision retrieval.",
  "Meanwhile, transformers continue to evolve with innovations like sparse attention and multimodal capabilities.",
  "Together, these components form the backbone of modern intelligent systems that blend reasoning, memory, and language fluency."
]
```

```python
len(document)
```

```python
query = "RAG systems"
```

```python
import re
def text_preprocessing(text):

  text = text.lower()
  clean_text = re.sub(r"[^\w\s]", "", text)

  return clean_text
```

```python
preprocessed_text = [text_preprocessing(doc) for doc in document]
preprocessed_text
```

```python
preprocessed_query = text_preprocessing(query)
preprocessed_query
```

```python
vector = TfidfVectorizer()
```

```python
X = vector.fit_transform(preprocessed_text)
```
